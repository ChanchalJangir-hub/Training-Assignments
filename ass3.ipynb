4. Create a class named Rectangle with attributes: 
o length  
o width  
Create a method to calculate and print the area of the rectangle. 
 

```python
class Rectangle:

    def __init__(self,length,width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width

        print("Area of Rectangle:-",area)

rect = Rectangle(13.0,7)

rect.calculate_area()
```

6.Create a class named BankAccount with: 
o account_holder  
o balance  
Create methods: 
o deposit(amount)  
o withdraw(amount)  
Display the updated balance after each transaction. 


```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)
        print("Updated Balance:", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
            print("Updated Balance:", self.balance)
        else:
            print("Insufficient Balance!")

account = BankAccount("John", 5000)

account.deposit(1000)
account.withdraw(2000)
account.withdraw(5000)
```

9. Create a class named Calculator with methods: 
o add()  
o subtract()  
o multiply()  
o divide()  
 Take two numbers as input and perform all operations.

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Cannot divide by zero"

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

calc = Calculator()

print("Addition:", calc.add(num1, num2))
print("Subtraction:", calc.subtract(num1, num2))
print("Multiplication:", calc.multiply(num1, num2))
print("Division:", calc.divide(num1, num2))
```

10. Create a class named LibraryBook with attributes:  
● book_name  
● author  
● price  
            Create a method display_book_info() that prints all book details. Create    
at least three book objects and display their information.

```python
class LibraryBook:
    def __init__(self, book_name, author, price):
        self.book_name = book_name
        self.author = author
        self.price = price

    def display_book_info(self):
        print("Book Name:", self.book_name)
        print("Author:", self.author)
        print("Price:", self.price)
        print("-" * 30)

book1 = LibraryBook("The Alchemist", "Paulo Coelho", 350)
book2 = LibraryBook("Harry Potter", "J.K. Rowling", 500)
book3 = LibraryBook("Wings of Fire", "A.P.J. Abdul Kalam", 250)

book1.display_book_info()
book2.display_book_info()
book3.display_book_info()
```
