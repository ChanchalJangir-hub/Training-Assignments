6. Create a list: 
arr = [10, 20, 30, 40, 50, 60] 
Write a function that takes the list as input and returns the list in 
reverse order without using the reverse() method.

```python
arr = [10, 20, 30, 40, 50, 60] 

def reverse_list(lst):
    return lst[::-1]

reverse_arr = reverse_list(arr)

print("Original list:-",arr)
print("Reverse_list:-",reverse_arr)

```

7. Create a tuple: 
data = (5, 10, 15, 20, 25, 30, 35) 
Write a program to: 
○ Count how many elements are divisible by 5 
○ Find the sum of all elements 
○ Find the average of the tuple 

```python
data = (5, 10, 15, 20, 25, 30, 35) 

count = 0

for num in data:
    if num%5 == 0:
        count += 1

total = sum(data)

avg = total / len(data)

print("Count of elements divisible by 5:", count)
print("Sum of all elements:", total)
print("Average of the tuple:", avg)
```

8. Create a dictionary: 
students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
} 
Write a program to: 
○ Find the student with the highest marks 
○ Find the student with the lowest marks 
○ Print only the students who scored more than 85 marks 

```python
students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
} 

highest_student =  max(students,key = students.get)
lowest_student = min(students,key = students.get)

print("Student that score highest marks:-",highest_student)
print("Student that score lowest marks:-",lowest_student)

print("Highest scorer:", highest_student, "-", students[highest_student])
print("Lowest scorer:", lowest_student, "-", students[lowest_student])

print("Students scoring more than 85 marks:")
for name, marks in students.items():
    if marks > 85:
        print(name, ":-", marks)
```

9. Write a function: 
count_frequency(arr) 
that takes a list as input and prints the frequency of each element. 
Example: [1, 2, 2, 3, 1, 4, 2] 
arr = 
 Output: 
 
1 -> 2 times 
2 -> 3 times 
3 -> 1 time 
4 -> 1 time

```python
def count_frequency(arr):
    freq = {}

    for item in arr:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    for key in freq:
        if freq[key] == 1:
            print(key,"1 ->",freq[key],"times")
        else:
            print(key,"1 ->",freq[key],"times")

arr = [1, 2, 2, 3, 1, 4, 2] 
count_frequency(arr)
```

10. Write a function: 
find_duplicates(arr) 
that takes a list as input and prints all duplicate elements present in 
the list. 
Example: 
arr = [10, 20, 30, 20, 40, 10, 50, 30] 
Output:
Duplicate elements are: 
10 
20 
30

```python
def find_duplicates(arr):
    duplicates = set()
    s = set()

    for item in arr:
        if item in s:
            duplicates.add(item)
        else:
            s.add(item)

    print("Duplicate elements are:")
    for item in duplicates:
        print(item)

arr = [10, 20, 30, 20, 40, 10, 50, 30]
find_duplicates(arr)
```
